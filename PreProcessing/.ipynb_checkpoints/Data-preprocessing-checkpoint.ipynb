```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA

# Generate a sample dataset
np.random.seed(42)

# Dates
date_rng = pd.date_range(start='1/1/2000', end='12/31/2022', freq='M')

# Countries
countries = ['USA', 'Canada', 'UK', 'Brazil']

# Features
features = ['Population', 'GDP', 'Oil_Reserves', 'Production']

# Create a DataFrame
data = pd.DataFrame(index=date_rng)
data['Country'] = np.random.choice(countries, size=len(date_rng))
data['Population'] = np.random.randint(1000000, 10000000, size=len(date_rng))
data['GDP'] = np.random.randint(100000, 1000000, size=len(date_rng))
data['Oil_Reserves'] = np.random.randint(1000, 10000, size=len(date_rng))
data['Production'] = np.random.randint(500, 5000, size=len(date_rng))

# Target variable (Oil Price)
data['Target_Price'] = np.random.uniform(50, 100, size=len(date_rng)) + \
                      10 * data['Production'] + \
                      5 * data['GDP'] + \
                      2 * data['Oil_Reserves']

# Display the sample data
print(data.head())

# ARIMA modeling example for a specific country (e.g., USA)
country_data = data[data['Country'] == 'USA']

# Fit ARIMA model
model = ARIMA(country_data['Target_Price'], order=(1, 1, 1))
result = model.fit()

# Forecast
forecast_steps = 12  # Adjust as needed
forecast = result.forecast(steps=forecast_steps)

# Display the forecast
print(forecast)
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA

# Function to fit ARIMA model and make predictions
def fit_arima_and_predict(data):
    model = ARIMA(data['Target_Price'], order=(1, 1, 1))
    result = model.fit()
    forecast_steps = 12  # Adjust as needed
    forecast = result.forecast(steps=forecast_steps)
    return forecast

# Generate a sample dataset
np.random.seed(42)

# Dates
date_rng = pd.date_range(start='1/1/2000', end='12/31/2022', freq='M')

# Countries
countries = ['USA', 'Canada', 'UK', 'Brazil']

# Features
features = ['Population', 'GDP', 'Oil_Reserves', 'Production']

# Create a DataFrame
data = pd.DataFrame(index=date_rng)
data['Country'] = np.random.choice(countries, size=len(date_rng))
data['Population'] = np.random.randint(1000000, 10000000, size=len(date_rng))
data['GDP'] = np.random.randint(100000, 1000000, size=len(date_rng))
data['Oil_Reserves'] = np.random.randint(1000, 10000, size=len(date_rng))
data['Production'] = np.random.randint(500, 5000, size=len(date_rng))

# Target variable (Oil Price)
data['Target_Price'] = np.random.uniform(50, 100, size=len(date_rng)) + \
                      10 * data['Production'] + \
                      5 * data['GDP'] + \
                      2 * data['Oil_Reserves']

# Display the sample data
print(data.head())

# Loop through each country and make predictions
for country in countries:
    country_data = data[data['Country'] == country]
    forecast = fit_arima_and_predict(country_data)
    
    # Display the results for each country
    print(f"\nForecast for {country}:")
    print(forecast)
```
