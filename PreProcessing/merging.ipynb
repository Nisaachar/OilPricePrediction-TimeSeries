```python
import pandas as pd
import os

# Path to the folder containing CSV files
folder_path = 'Processed Data'  # Update with your folder path

# Get a list of all CSV files in the folder
all_files = os.listdir(folder_path)
csv_files = [file for file in all_files if file.endswith('.csv')]

# Initialize an empty DataFrame to store the merged data
merged_df = pd.DataFrame()

# Loop through each CSV file and merge based on 'year' and 'country' columns
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    if merged_df.empty:
        merged_df = df
    else:
        merged_df = pd.merge(merged_df, df, on=['Year', 'Country'])

# Write the merged DataFrame to a new CSV file
merged_file_path = 'merged_data.csv'  # Update with your desired output file path
merged_df.to_csv(merged_file_path, index=False)

# Display a message when the process is complete
print(f"Merged data saved to {merged_file_path}")

merged_df.head()
```
